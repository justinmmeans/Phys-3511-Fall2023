1. Define a function called **MakeHist** that will make a histogram and takes as its arguments a numpy array (data), integer (number of bins), and strings for its title and horizontal and vertical axis titles. 


2. Once you have your MakeHist function defined, create a numpy array generated from the random normal distribution centered at zero. 

3. Call your MakeHist function 3 times with bin values of 10, 100, and 1000. Each call should use your random numpy array as the input data. For this exersice you can select any name for the horizontal and vertical axes, but set the histogram title to the number of bins it was produced with.

Expected Result (note your specific histogram numbers will differ since they are generated randomly, but the shape should be the same):

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
#%matplotlib inline


def makeHist(array, binsAMT, horizTitle, vertTitle, histTitle): #function to generate a histogram
    
    histogramGen = plt.figure()
    axes = histogramGen.add_axes([.1,.1,.8,.8]) #size of table
    axes.set_xlabel(horizTitle) #names x axis title
    axes.set_ylabel(vertTitle) #names y axis title
    axes.set_title(histTitle) #names overall title of dataset
    axes.hist(array,binsAMT) #generates histogram based on input array and amt of bins input by user
    
newArr = np.random.randn(1000) #generate array of random integers
```

```python
makeHist(newArr, 10, 'X Data Set Title', 'Y Data Set Title', 'Data with 10 Bins')
```

```python
makeHist(newArr, 100, 'X Data Set Title', 'Y Data Set Title', 'Data with 100 Bins')
```

```python
makeHist(newArr, 1000, 'X Data Set Title', 'Y Data Set Title', 'Data with 1000 Bins')
```
